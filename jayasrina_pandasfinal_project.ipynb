```python
import pandas as pd
df = pd.read_excel("/content/mega_data_cleaning_dataset.xlsx")
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.drop_duplicates(subset='Customer_ID', inplace=True)
df
```

```python
df.drop_duplicates(subset='Email', inplace=True)
df
```

```python
df.drop_duplicates(subset='Product_ID', inplace=True)
df
```

```python
df['Name'].fillna('Unknown', inplace=True)
```

```python
df['Email'].fillna('noemail@example.com', inplace=True)
```

```python
df['Product_Name'].fillna('Misc', inplace=True)
```

```python
df['Category'].fillna('Misc', inplace=True)
```

```python
df['Feedback_Score'] = df.groupby('Product_ID')['Feedback_Score'].transform(lambda x: x.fillna(x.median()))
df
```

```python
df['Quantity'].fillna(1, inplace=True)

```

```python
df = df[df['Price'] > 0]
df
```

```python
df = df[df['Quantity'] >= 0]
```

```python
df = df[(df['Age'] >= 0) & (df['Age'] <= 120)]
df
```

```python
df = df[(df['Feedback_Score'] >= 1) & (df['Feedback_Score'] <= 5)]
df
```

```python
df['Delivery_Date'] = pd.to_datetime(df['Delivery_Date'])
```

```python
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])
```

```python
df = df[df['Delivery_Date'] >= df['Purchase_Date']]
df
```

```python
df['Gender'] = df['Gender'].replace({'Mle': 'Male', 'f': 'Female'})
df
```

```python
df['Payment_Type'] = df['Payment_Type'].str.strip().str.title()
df
```

```python
df['Account_Status'] = df['Account_Status'].str.strip().str.title()
df
```

```python
df = df[df['Email'].str.contains(r'^[\w\.-]+@[\w\.-]+\.\w+$', na=False)]
df
```

```python
df[df['Price'] > 500]
```

```python
df[df['Feedback_Score'] < 3]
```

```python
df[(df['Customer_Loyalty'] == 'VIP') & (df['Total_Purchase'] > 1000)]
```

```python
df[df['Account_Status'] == 'Banned']
```

```python
df[(df['Delivery_Date'] - df['Purchase_Date']).dt.days > 7]
```

```python
df[df['Category'] == 'Electronics']
```

```python
df[df['Country'] == 'India']
```

```python
df[df['Quantity'] > 10]
```

```python
df[df['Discount'] > 50]
```

```python
df[df['Price'] * df['Quantity'] > 1000]
```

```python
df['Total_Purchase']=df['Price']*df['Quantity']
df
```

```python
df['discounted_price']=df['Price']-df['Discount']
df
```

```python
df['Delivery_Duration'] = (df['Delivery_Date'] - df['Purchase_Date']).dt.days
df
```

```python
def age_group(age):
  if age<13:
    return 'child'
  elif age<20:
    return 'teen'
  elif age<60:
    return 'adult'
  else:
    return 'senior'
df['age_group']=df['Age'].apply(age_group)
df
```

```python
def loyalty(total):
    if total > 50000:
        return 'Gold'
    elif total > 20000:
        return 'Silver'
    else:
        return 'Bronze'

df['Loyalty_Level'] = df['Total_Purchase'].apply(loyalty)
df
```

```python
df['gender_code']=df['Gender'].map({'Male':1,'Mle':1,'Female':2,'Other':3})
df
```

```python
df['Account_Status_code']=df['Account_Status'].map({'Active':1,'Inactive':0,'Banned':2})
df
```

```python
df['Country']=df['Country'].str.strip().str.title()
df
```

```python
df['Payment_Type']=df['Payment_Type'].str.strip().str.title()
df
```

```python
df=df.apply(lambda x:x.str.strip().str.lower()if x.dtype=="object" else x)
df
```

```python
df['Revenue'] = df['Price'] * df['Quantity']
top5_products = df.groupby('Product_Name')['Revenue'].sum().sort_values(ascending=False).head(5)
print(top5_products)
```

```python
top5_customers = df.groupby('Customer_ID')['Total_Purchase'].sum().sort_values(ascending=False).head(5)
print(top5_customers)
```

```python
avg_discount = df.groupby('Category')['Discount'].mean()
print(avg_discount)
```

```python
avg_feedback_cat = df.groupby('Category')['Feedback_Score'].mean()
```

```python
popular_payment = df['Payment_Type'].value_counts().idxmax()
print("Most popular payment type:", popular_payment)
```

```python
df['Order_Value'] = df['Price'] * df['Quantity']
top_customer = df.loc[df['Order_Value'].idxmax(), ['Customer_ID', 'Order_Value']]
print(top_customer)
```

```python
frequent_customers = df['Customer_ID'].value_counts()
frequent_customers = frequent_customers[frequent_customers > 5]
print(frequent_customers)
```

```python
sorted_delivery = df.sort_values(by='Delivery_Duration', ascending=False)
df
```

```python
max_feedback = df.loc[df['Feedback_Score'].idxmax()]
min_feedback = df.loc[df['Feedback_Score'].idxmin()]
df
```

```python
grouped_purchase = df.groupby(['Country','Category'])['Total_Purchase'].sum().reset_index()
df
```
